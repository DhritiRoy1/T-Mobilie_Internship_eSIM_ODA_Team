# Removed Unnecessary Things&Tokenized Tweet

```python
Review=[]
FilteredReviews=[[]]
#WordCount=1
for h in range(len(Text)):
    #splits the words in each line to a list of words
    Tweet =Text.iloc[h,1]
    Tweet=str(Tweet).lower()
    Tweet_Without_Hyperlink = re.sub(r'http\S+', '', Tweet)
    FinalTweet= re.sub(r'[^\w\s]', '', Tweet_Without_Hyperlink)
    FinalTweet=FinalTweet.replace('\r', '')
    FinalTweet=FinalTweet.replace('\n', '')
    Review = FinalTweet.split(' ')
    for word in Review:
        Review[Review.index(word)]= lemmatizer.lemmatize(word)
    #print(Review)
    #print(len(Review))
    for j in Review:
        if j not in StopWords:
            FilteredReviews[h].append(j)
    FilteredReviews.append([])

#del FilteredReviews[h]
    #print(FilteredReviews[h])
print(FilteredReviews)

```
